```python
import spacy
from spacy import displacy
```

```python
nlp2 = spacy.load('C:\\Users\\applied informatics\\pipVirtualEnvironments\\nlpVenv\\workContents')
```

```python
import pickle
```

```python
with open('./spaCyTestDataNonEmptyEntities.txt', 'rb') as pf:
    TEST_DATA = pickle.load(pf)
```

```python
for text, _ in TEST_DATA:
    doc = nlp2(text)
    displacy.render(doc,style="ent")
```

```python
from spacy.gold import GoldParse
from spacy.scorer import Scorer
```

```python
def evaluate():
    scorer = Scorer()
    for text, annot in TEST_DATA:
        doc_gold_text = nlp2.make_doc(text)
        gold = GoldParse(doc_gold_text, entities=annot['entities'])
        pred_value = nlp2(text)
        scorer.score(pred_value, gold)
    print(scorer.scores)
```

```python
evaluate()
```
